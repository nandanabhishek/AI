```python
# Step 1 : Load Data
import pandas as pd
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset['Level']
print(X)
```

```python
y = dataset['Salary']
print(y)
```

```python
# Step 2 : Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Python scikit-learn only accepts the training and test data in a 2-dimensional array format. 
# We have to perform data transformation on our training set and test set.
import numpy as np

X_train = np.array(X_train).reshape((len(X_train),1))
print(X_train)
```

```python
y_train = np.array(y_train).reshape((len(y_train),1))
print(y_train)
```

```python
X_test = np.array(X_test).reshape(len(X_test), 1)
print(X_test)
```

```python
y_test = np.array(y_test).reshape(len(y_test), 1)
print(y_test)
```

```python
# Step 3 : Fit Simple Linear Regression to Training Data
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Step 4 : Make Prediction
y_pred = regressor.predict(X_test)
```

```python
# Step 5 : Visualize training set results
import matplotlib.pyplot as plt
# plot the actual data points of training set
plt.scatter(X_train, y_train, color = 'red')
# plot the regression line
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

print('\n')
```

```python
# Step 6 : Visualize test set results
import matplotlib.pyplot as plt
# plot the actual data points of test set
plt.scatter(X_test, y_test, color = 'red')
# plot the regression line (same as above)
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python

# Step 7 : Make new predictions
new_salary_pred1 = regressor.predict([[6.5]])
print('\n\nThe predicted salary of a person with 6.5 years experience is ', new_salary_pred1)
new_salary_pred2 = regressor.predict([[7.9]])
print('The predicted salary of a person with 7.9 years experience is ', new_salary_pred2)
new_salary_pred3 = regressor.predict([[20]])
print('The predicted salary of a person with 20 years experience is ', new_salary_pred3)
```
