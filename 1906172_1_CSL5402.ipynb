# Question-1 : Program to implement simple calculator

```python
# python program for implementation of Simple Calculator

# function for addition operation
def add(a, b):
    return a+b

# function for subtraction operation
def sub(a, b):
    return a-b

# function for multiplication operation
def multiply(a, b):
    return a*b

# function for division operation
def divide(a, b):
    if b != 0:
        return a/b
    return -1


while True:
    print('Select Operation.')
    print('1. Add')
    print('2. Subtract')
    print('3. Multiply')
    print('4. Divide')


    # Take input from the user
    choice = input('\nEnter choice (1/2/3/4) : ')

    # Check if choice is one of the four options
    if choice in ('1', '2', '3', '4'):
        num1 = float(input('\nEnter first number : '))
        num2 = float(input('Enter second number : '))

        if choice == '1':
            print('{} + {} = {}'.format(num1, num2, add(num1, num2)))

        elif choice == '2':
            print('{} - {} = {}'.format(num1, num2, sub(num1, num2)))

        elif choice == '3':
            print('{} * {} = {}'.format(num1, num2, multiply(num1, num2)))

        elif choice == '4':
            if num2 == 0:
                print('Division operation is undefined when denominator is 0 !')
            else:
                print('{} / {} = {}'.format(num1, num2, divide(num1, num2)))

    else:
        print("Invalid Input !")
        
    print('\n')
    option = int(input('Enter 1 to use the calculator again, 0 to exit : '))
    if option == 1:
        continue
    else:
        break
```

# Question-2 : Program to check whether an year is Leap year or not 

```python
# python program to check whether an year is Leap year or not


while True:
    year = int(input('Enter an year : '))
    if year <= 0:
        print('Invalid year value !')
    else:
        if year % 4 == 0:
            if year % 100 == 0:
                if year % 400 == 0:
                    print(year, 'is a Leap Year.')
                else:
                    print(year, 'is not a Leap Year.')
            else:
                print(year, 'is a Leap Year.')               
        else:
            print(year, 'is not a Leap Year.')
            
    print('\n')
    option = int(input('Enter 1 to use this program again, 0 to exit : '))
    if option == 1:
        continue
    else:
        break      
```

# Question-4 : Program to shuffle a deck of cards

```python
# Python program to shuffle a deck of card

# importing modules
import itertools, random

while True:
    # make a deck of cards
    deck = list(itertools.product(range(1,14), ['Spade', 'Club', 'Heart', 'Diamond']))

    # shuffle the cards
    random.shuffle(deck)

    # draw n cards randomly
    cards = int(input('Enter the number of cards you wish to draw : '))
    print('You got :')
    for i in range(cards):
       print(deck[i][0], 'of', deck[i][1])
    
    print('\n')
    option = int(input('Enter 1 to use this program again, 0 to exit : '))
    if option == 1:
        continue
    else:
        break
```
