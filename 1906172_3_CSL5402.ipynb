```python
import pandas as pd

df = pd.read_csv('/content/employees.csv')

# to print information about the data
print("\n\nInformation of employee.csv original file :\n")
df.info()

# to check number of null values present in each column
print('\n\nInformation about number of NULL values present in each column (in original employee.csv) :\n')
print(df.isnull().sum())

# to remove all the null values in data
new_df = df.dropna()

# to print information about the data
print("\n\nInformation of employee.csv file after removing NULL values :\n")
new_df.info()

# to check number of null values present in each column
print('\n\nInformation about number of NULL values present in each column after modification :\n')
print(new_df.isnull().sum())

# printing whole modified data
print('\n\n Modified dataset :\n')
print(new_df.to_string())


# to sort dataset in descending order
sorted_df = new_df.sort_values('Salary', ascending=False)
print('\n\nSorted dataset by Salary in descending order :\n')
print(sorted_df.to_string(), '\n\n')
sorted_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/employees.csv')

# histogram

# scatter graph
x = df['Salary']
y = df['Bonus %']

plt.figure(figsize=(12,10))
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.xlabel("Salary",  fontsize= 15, color='green')
plt.ylabel("Bonus %",  fontsize= 15, color='green')
plt.title('Scatter Graph', fontsize=20, color='blue')
plt.scatter(x, y, color='red')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/employees.csv')

# line graph
x = df['Salary']
y = df['Bonus %']

plt.figure(figsize=(14,10))
plt.grid(color = 'lightgreen', linestyle = '--', linewidth = 0.5)
plt.xlabel("Salary",  fontsize= 15, color='green')
plt.ylabel("Bonus %",  fontsize= 15, color='green')
plt.title('Line Graph', fontsize=20, color='blue')

plt.plot(x, y, color='red', label='Line Graph')
plt.legend(loc='best')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/employees.csv')

# Histogram - A
x = df['Salary']

plt.figure(figsize=(14,10))
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)

plt.xlabel('Salary', fontsize=14, color='green')
plt.ylabel('No. of Employees', fontsize=14, color='green')
plt.title('Histogram-Salary', fontsize=15, color='blue')
plt.hist(x, color='orange', label='Histogram')
plt.legend(loc='best')
plt.show()


# Histogram - B
y = df['Bonus %']

plt.figure(figsize=(14,10))
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)

plt.xlabel('Bonus %', fontsize=14, color='green')
plt.ylabel('No. of Employees', fontsize=14, color='green')
plt.title('Histogram-Bonus %', fontsize=15, color='blue')
plt.hist(y, color='hotpink', label='Histogram')
plt.legend(loc='best')
plt.show()
```
